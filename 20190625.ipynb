```
#reindex 
obj = Series([2.5, 6.4, 7.8, 6.5], index = ["kim", "cho", "bae", "lee"])
print(obj,"\n")
obj1 = obj.reindex(["lee", "kim", "bae", "cho","choi"])
print(obj1)

obj2 = obj.reindex(["bae", "cho", "choi", "kang", "kim", "lee"], fill_value= -50)
print(obj2)

obj3 = obj.reindex(["bae", "cho", "choi", "kang", "kim", "lee"], method= "bfill")
print(obj)
```

```
#DataFrame의 reindex
data = {"mid" : [30, 32, 28, 20, 19], "hw" : [10, 9, 7, 8, 10], "fin" : [29, 27, 31, 18, 21],
       "gender" : ["M", "F", "M", "M", "F"]}
df = DataFrame(data, index=[2017103, 2017204, 2018212, 2018217, 2019137])
df
```

```
#2107103,  2017204, 2018212, 2018217, 2019108, 2019137 <-reindex, Nan은0으로 채우기
#열 index에 reindex를 써서 "att"인덱스 추가
#행 indec에 "ffill" method를 사용해서 결과값홧인 : 두 줄 위의 코드 사용
df1 = df.reindex([2107103,  2017204, 2018212, 2018217, 2019108, 2019137], fill_value = 0)
print(df1)
print("\n")

df2 = df1.reindex([2107103,  2017204, 2018212, 2018217, 2019108, 2019137], method = "ffill")
print(df2)
print("\n")

df3 = df.reindex(columns=["mid", "hw", "fin", "gender", "att"], fill_value = -1)
print(df3)
```

```
#drop

df1 = df.drop(2018212)
print(df1)
print("\n")




#여러 개의 행을 drop하려면 drop할 행을 리스트로 만들어서 drop호출
#2017204, 2018212, 2019137 drop
d = Series([2017204, 2018212, 2019137])
df2 = df.drop(d)
print(df2)
print("\n")

#열을 drop하려면 axix=1만 추가
df3 = df.drop(["mid", "hw"],axis=1) 
print(df3)
print("\n")

#df2에서 att, hw를 drop
df4 = df.drop(columns="att")
print(df4)
print("\n")
```

```
#계층색인
s = Series(np.random.binomial(n =10, p=0.4, size = 10),
          index = [["stat", "stat", "stat", "econ", "econ",
                   "comp", "comp", "comp", "math", "math"],
                  [1, 2, 3,1, 2, 2, 3, 4, 3, 4 ]])
print(s)
```

```
#Series 계층적 인덱싱
print(s["comp"])
print(s[["stat", "math"]])
print("\n")
print(s[["stat", "math"]][:2])
```

```
#중첩
print(s)
print("\n")
print(s.unstack()) #2d index를 column으로 올림
print("\n")
print(s.unstack().stack())
```
