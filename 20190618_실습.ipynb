```
def fib(n):
    a, b = 0, 1
    while a < n:
      print(a, end=' ')
      a, b = b, a+b
    print()
fib(1000)
```

**continue는 조건이 맞으면 다시 조건문으로 돌아가서 반복하라**

```
for i in range(3):
  a=int (input("더하고자 하는 첫 번째 값을 입력하세요.(정수) : "))
  b=int (input("더하고자 하는 두 번째 값을 입력하세요.(정수) : "))
  if(a==0 and b==0):
    continue
  else:
    break
  print("{0} + {1} = {2}".format(a,b,a+b))
  
  print("계산을 종료합니다.")
```

**Literal의 개념**

**tuple은 바꾸지 못한다.**

**list는 내부로 들어가보면 바꿀 수 있음**

**딕셔너리에서 {key,value,  key,value}로 구성된다**



```
band="The doors" #리스트
Ls1 = list(band)
print(Ls1)

tp=(25,33,47,5) #튜플
Ls2=list(tp)
print(Ls2)

st={25,13,45,65} #집합
Ls3=list(st)
print(Ls3)

dic={"서울" : "02",  "경기" : "031",  "제주" : "064"} #딕셔너리
Ls4 =list(dic)
print(Ls4)

ran= range(1,10,3)  #range
Ls5 = list(ran)
print(Ls5)

rev = reversed([1,2,3,4]) #reversed
Ls6 = list(rev)
print(Ls6)
```
