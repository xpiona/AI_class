```
str="python program"       #문자열
ls = [25,42,82,76,91,44]   #리스트
tp = (5,3,9,8,1,4)         #튜플
st={100,200,300,400,105}   #집합
dic = {"서울" : "02", "경기" : "031", "강원" : "033", "세종" : "044"} #사전

#데이터가 겹치지 않게 6개 뽑음
print(random.sample(str,6))
print(random.sample(ls,3))
print(random.sample(tp,3))

print(random.sample(st,3))
#print(random.sample(dic,3))

list(dic)
print(random.sample(list(dic),3))
```

```
import itertools
dir(itertools)

str = "python"
ls = [25,42,82,76,91,44]   #리스트
dic = {"서울" : "02", "경기" : "031", "강원" : "033", "세종" : "044"} #사전

#print(list(itertools.chain(str,ls,dic)))
for x in itertools.chain(ls,str,dic):
  print(x, end= "\t")
print()

for x in itertools.count():
  if x>20 :
    break
  print(x,end="\t")
print()

list_alpha = ["A", "B", "C", "D", "E", "F"]
list(zip(itertools.count(), list_alpha))
```
