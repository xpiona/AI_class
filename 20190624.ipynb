```
#comma가 없다는것은 리스트가 아니다.
import numpy as np
ls = [2,6,8,7,5,6]
x =  np.array([2,6,8,7,5,6])
print(ls[2],ls[:4],ls[1:4:2])
print(x[2], x[:4], x[1:4:2])

ls[2:5] = [6] #2번부터 5번전까지의 리스트를 [6]으로 변경
print(ls) #mutable data
#x[2:5] =6
print(x)
```

```
x_sub1 = x[1:5]
x_sub2 = x[1:5].copy()

print(x, x_sub1, x_sub2)

x[2] = 100

print(x, x_sub1, x_sub2)

```

```
name = np.array(["남자", "여자", "남자", "남자", "남자", "여자"])
score = np.array([[10,20,30,40],[8,15,24,31],[5,11,20,25],
                [6,9,16,17],[10,17,21,27],[9,13,27,31]])
name_bool = name =="여자"

print(name,"\n","\n",score,"\n","\n",name_bool)
```

```
female = score[name_bool]
print(female)

print("\n")

male = score[~name_bool]
print(male)
```

```
cls = np.array(["1학년", "2학년","3학년","1학년", "2학년","3학년"])

score = np.array([[10,20,30,40],[8,15,24,31],[5,11,20,25],
                  [6,7,16,17],[10,17,21,27],[9,13,27,31]])

print(score[(cls=="1학년") | (cls == "2학년")])
```

```
score[(cls=="1학년") | (cls == "2학년")] = 0
print(score)
```

```
score[score >= 20] =100
print(score)
```

```
arr = np.random.rand(7,4)
print(arr)
```

```
arr1 = arr[[6,3,5]] #6번행 3번행 5번행 차례대로 가져와라 이게 펜스 인덱싱
print(arr1)
```

```
arr2 = arr[[-1,-4,-2]] #맨 마지막 행 부터 0, -1, -2 순서
print(arr2)
```

```
arr3 = arr[[5,0,2],[1,0,2]] #5번,0번,2번행에서 (5번행에서 가로로 1번째 값, 0번행에서 가로로 0번째 값을 가져온다.
print(arr3)
```

```
arr4 = arr[[5,0,2]][:,[1,0,2]] #5,0,2행에서 1번원소,0번원소,2번원소를 차례대로 뽑아옴
arr5 = arr[np.ix_([5,0,2],[1,0,2])]

print(arr4)
print(arr5)
```

```
arr = np.arange(28).reshape(4,7)
print(arr)
print("\n")
arr1 = arr.T
arr2 = arr.transpose()
print(arr1)
print("\n")
print(arr2)
```

```
arr = np.arange(24).reshape(2,4,3) #앞에서부터 인덱스가 0,1,2임
print(arr)
print("\n")
arr1 = arr.transpose(2,0,1) #위의 인덱스를 이용 인덱스 번호에 따라서 3,2,4가 들어감
print(arr1)
```

```
x = np.array([2.5, 5.6, 8.1, 3.9, 5.7])
y = np.array([2, 6, -3, 4, -5])
A = np.array([x,y])

print(np.sqrt(x))
print("\n")
print(np.exp(x))
print("\n")
print(np.rint(np.exp(x))) #소수점 자르기
print("\n")
print(np.log(x))
print("\n")
print(np.abs(y))
print("\n")
print(np.sign(y))
print("\n")
r, q = np.modf(A) #실수부봐 소수부로 찢음
print(r)
print("\n")
print(q)
```
