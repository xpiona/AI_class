```
xarr = np.array([2.5, 5.6, 8.1, 3.9, 5.7]) 
print("x.sum() = {0}\tsum(x) = {1}".format(xarr.sum(), np.sum(xarr)))
print("x.mean() = {0}\tmean(x) = {1}".format(xarr.mean(), np.mean(xarr)))
print("x.std() = {0}\tstd(x) = {1}".format(xarr.std(), np.std(xarr)))

print("x.cumsum() = {0}\tcumsum(x) = {1}".format(xarr.cumsum(), np.cumsum(xarr))) #accumulated sum 누적합
```

```
xarr = np.array([[2.5, 5.6, 8.1], [3.9, 5.7, 2.9], [6.1, 9.7, 5.4],[1.4, 8.3, 4.2]])
print(xarr)
print("\n")
print("세로 평균\n{0}\n가로평균\n{1}".format(xarr.mean(axis=0),xarr.mean(axis=1)))
print("\n")
print("전체 최소값의 인덱스 : {0}".format(xarr.argmin()))
print("세로(열) 최소값의 인덱스 : {0}".format(xarr.argmin(axis=0)))
print("가로(행) 최소값의 인덱스 : {0}".format(xarr.argmin(axis=1)))
print("\n")
print("전체 누적합 : {0}".format(xarr.cumsum()))
print("열 누적합 : \n{0}".format(xarr.cumsum(axis=0)))
print("행 누적합 : \n{0}".format(xarr.cumsum(axis=1)))
```

```
A = np.array([[1,3,5,7],[5,3,4,7]])
B = np.array([[1,3],[5,7],[5,3],[4,7]])
C = np.array([[4,9,1,3],[7,5,2,7],[5,3,4,3],[1,4,7,9]])

print(A)
print(B)
print(C)
```

```
print("AB\n{0}".format(np.dot(A,B)))
print("C행렬의 대각원소 : {0}".format(np.diag(C)))
print("\n")
print("A행렬의 1행렬을 대각원소로 하는 행렬 \n: {0}".format(np.diag(A[1])))
print("C행렬의 대각원소의 합 \n: {0}".format(np.trace(C)))
print("C행렬의 행렬식 \n{0}".format(np.linalg.det(C)))
```

```
print("C행렬의 행렬식 \n{0}".format(np.linalg.inv(C)))
```

```
eig_v,eig_vec = np.linalg.eig(C)
print("eigenvalue \n{0}\neigenvector\n{0}".format(eig_v,eig_vec))
```
